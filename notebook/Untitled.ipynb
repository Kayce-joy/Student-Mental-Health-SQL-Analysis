```python
# Step 1: Import libraries
import pandas as pd
from sqlalchemy import create_engine

# Step 2: Load CSV
csv_file = 'Stress Indicators Dataset for Mental Health Classification.csv'
df = pd.read_csv(csv_file)

# Preview first 5 rows
df.head()

# Step 3: Create a SQLite database file
engine = create_engine('sqlite:///stress_dataset.db')  # creates file database

# Load CSV into SQL table called 'stress'
df.to_sql('stress', con=engine, index=False, if_exists='replace')
print("CSV loaded into SQL table: 'stress'")
```
